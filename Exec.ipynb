# Ballot Measures Funding Local Government
This data set shows the success rate of local government funding measures across the state of Wahsington from 2011-2017. The goal of this research is to find patterns in what voters are willing to approove to expand local government in light of property tax caps in state law.

## Descriptive Statistics
Before any analysis is possible, we have to understand the general patterns in the data. To understand the data it is important to know that it takes a majority vote to raise taxes in a district. As the histogram below shows, only a small proportion of measures fail, while the average meaure revceives about 60% yes votes.

```python
import pandas as pd
data= 'https://github.com/nbennett4025/govproject/raw/master/ballotinitiativesWA.xlsx'
ballot=pd.read_excel(data,0)
ballotyes=ballot['Yes %']
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(ballotyes.dropna(),kde=False,fit=norm)
plt.legend(('data fit to normal','data as it is'))
plt.title('How often do funding measures get 50% in WA?')
plt.show()
```

## Breakdown by government type, timing, and Subject
There are a lot of different governments captured within the data. To facilitate analysis we can simplify the large number of categories into three big ones: cities, counties, and various special purpose governments. Counties, which proivde both regional and local services, have the most difficult time getting funding approved.

```python
values = list(ballot['Government Type'].value_counts().index)
city = values[0]
county = values[3]
govsimple = []

for value in ballot['Government Type']:
    if value in city:
        govsimple.append('City')
    else:
        if value in county:
            govsimple.append('County')
        else:
            govsimple.append('Special District')
            
ballot['govsimple'] = govsimple
sns.factorplot(x='govsimple', y= 'Yes %', data=ballot)
plt.title('Approval Rate by Government Type')
plt.show()
```

```python
sns.factorplot(x='Election Month', y= 'Yes %', data=ballot, size=5)
plt.title('Approval Rate by Election Month')
plt.show()
```

```python
import numpy as np
subjects = list(ballot['Subject'].value_counts().index)
parks = subjects[1]
transit = subjects[3]
govorg =subjects[5]
cemetery=subjects[7]
other=subjects[4]
water=subjects[12]
publicsafety = ['EMS', 'Fire', 'Fire, EMS', 'Criminal Justice']
human = ['Hospital', 'Library', 'Affordable Housing', 'Human Services']
annexation = ['Annexation, Fire', 'Annexation', 'Annexation, Library', 'Annexation, Transportation', 'Annexation, Water-Sewer']
subsimple = []
for value in ballot['Subject']:
    if value in parks:
        subsimple.append('Parks')
    else:
        if value in transit:
            subsimple.append('Transit')
        else:
            if value in govorg:
                subsimple.append('Government Organization')
            else: 
                if value in cemetery:
                    subsimple.append('Cemetery')
                else:
                    if value in water:
                        subsimple.append('Water/Sewer')
                    else:
                        if value in other:
                            subsimple.append('Other')
                        else:
                            if value in publicsafety:
                                subsimple.append('Public Safety')
                            else:
                                if value in human:
                                    subsimple.append('Human Services')
                                else:
                                    if value in annexation:
                                        subsimple.append('Annexation')
                                    else:
                                        subsimple.append('Multiple')
                                        
ballot['subsimple'] = subsimple
sns.factorplot(x='subsimple', y= 'Yes %', data=ballot,size=7, aspect=2)
plt.title('Approval Rate By subject matter')
plt.show()
```

Each of these factors shows significant variation in support. County governments fare worse than cities or special districts. For election timing it seems that any non-November election is more likely to yield positive results. 

We can also see fairly large variations in how different subjects fare on the ballot. In general local services like a cemetaries, water/sewer, and public safety do well. Regional services like transit and government itself are less likely to do well. Annexation is tricky, becuase unlike other issues covered here it requires a larger than 50% margin to pass. As a consequence annexations are much less likely to pass even though they have a higher average support.

## Mapping Funding Failures
While subject, government type, and timing of elections may explain variations in passage, spatial information might also provide some insight. This map shows which counties have the least success in passing funding measures.

```python
import geopandas as gpd
compressedmap = 'C:/Users/benne/Desktop/govproject/govproject/shape.shp'
countymap =gpd.GeoDataFrame.from_file(compressedmap)
faildata = 'https://github.com/nbennett4025/govproject/raw/master/countyballotfail.xlsx'
ballotfail = pd.read_excel(faildata,0)
mergedfailmap=countymap.merge(ballotfail, left_on='name', right_on='County')
base = mergedfailmap.plot(color='black', figsize=(20,13))
topLayer=mergedfailmap.plot(ax=base,column='failpercent',cmap='Reds',scheme='Quantiles',k=5,legend=True)
topLayer.set_title('Failure rate of government funding measures in Washington by County (2011-2017)',
                      color='black',fontdict={'fontsize':30})
plt.show()
```

This map of the % of ballot measures that fail does not show a clear regional pattern but it does hel us decontruct stereotypes about the state. While many assume that the easturn part of the state is more conservative and anti-government, there are many counties in Eastern Washington where ballot measures are more likely to pass than the liberal hubs of Western Washington.

While we are able to see clear differences in the performance of funding measures on ballots throughout the state, it is difficult from this data to explain differences in performance. Further research is required to build a workable model of what effects passage rates.
